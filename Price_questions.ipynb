## Clean

```python
# Remove $ and , and convert the columns to float to ease analysis
sea.price=sea.price.str.replace('$','')
sea.price=sea.price.str.replace(',','').astype(float)
bos.price=bos.price.str.replace('$','')
bos.price=bos.price.str.replace(',','').astype(float)
```

## Analyze and model

### What is the average , maximum and minimum price in Boston and Seattle ?

```python
#Average price in Boston
average_bos=bos['price'].mean()

#Average price in Seattle
average_sea=sea['price'].mean()

# Minimum Price in Seattle
min_seattle = sea['price'].min()

# Minimum Price in Boston
min_bos = bos['price'].min()

#Maximum price in Boston
max_bos = bos['price'].max()

# Maximum price in Seattle
max_sea = sea['price'].max()
```

### What is the average rating for prices greater than 500 , between 250  and 500 , and for less than 250 dollars in boston and seattle?

```python
#Average rating for prices between 250$ and 500$ in Boston
average_bos_250_500 = bos['review_scores_rating'][(bos['price']>250) & (bos['price']<500)].mean()

#Average rating for prices  between 250$ and 500$ in Seattle
average_sea_250_500 = sea['review_scores_rating'][(sea['price']>250) & (sea['price']<500)].mean()

#Average rating for prices less than 250$ in Seattle
average_sea_250 = sea['review_scores_rating'][sea['price']<250].mean()

#Average rating for prices less than 250$ in Boston
average_bos_250 = bos['review_scores_rating'][bos['price']<250].mean()

#Average rating for prices greater than 500$ in Seattle
average_sea_500 = sea['review_scores_rating'][sea['price']>500].mean()

#Average rating for prices greater than 500$ in Boston
average_bos_500 = bos['review_scores_rating'][bos['price']>500].mean()
```
